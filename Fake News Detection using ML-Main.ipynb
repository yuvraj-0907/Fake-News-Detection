```python
merge=pd.concat([fake,true],axis=0)
merge.head(10)
```

```python
df=merge.drop(["title","subject","date"],axis=1)
df.head(10)
```

```python
df=df.sample(frac=1)
```

```python
import re
import string
def word_drop(text):
    text=text.lower()
    text=re.sub('\[.*/\]','',text)
    text=re.sub("\\W"," ",text)
    text=re.sub('https?://\S+|www.\.\S+','',text)
    text=re.sub('<.*?>+','',text)
    text=re.sub('[%s]'%re.escape(string.punctuation),'',text)
    text=re.sub('\n','',text)
    text=re.sub('\w*\d\w*','',text)
    return text
```

```python
df["text"]=df["text"].apply(word_drop)
```

```python
df.head(10)
```

```python
x=df["text"]
y=df["class"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size= .25)
```

```python
vectrorization=TfidfVectorizer()
xv_train=vectrorization.fit_transform(x_train)
xv_test=vectrorization.transform(x_test)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
LR.score(xv_test,y_test)
```

```python
pred_LR=LR.predict(xv_test)
```

```python
print(classification_report(y_test,pred_LR))
```

#### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier  
```

```python
DT=DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
DT.score(xv_test,y_test)
```

```python
pred_DT=DT.predict(xv_test)
```

```python
print(classification_report(y_test,pred_DT))
```

#### Manual Testing

```python
import pandas as pd
import numpy as np

def output_label(n):
    """ Converts prediction to a human-readable label. """
    if isinstance(n, (list, np.ndarray)):
        return [output_label(x) for x in n]
    return "Fake News" if n == 0 else "Not a Fake News"

def manual_testing(news):
    # Ensure that the news string is properly formatted
    news = news.strip()  # Remove any leading or trailing whitespace
    
    # Create DataFrame from news input
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    
    # Apply preprocessing to the text data
    new_def_test["text"] = new_def_test["text"].apply(word_drop)
    new_x_test = new_def_test["text"]
    
    # Vectorize the test data
    new_xv_test = vectrorization.transform(new_x_test)
    
    # Predict using both models
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    
    # Print predictions
    print("\n\nLogistic Regression Prediction: {} \nDecision Tree Prediction: {}".format(
        output_label(pred_LR[0]),  # Assuming pred_LR and pred_DT are arrays
        output_label(pred_DT[0])
    ))

# Example usage
news = input("Enter the news: ")
manual_testing(news)


```
